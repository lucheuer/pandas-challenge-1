## Part 2: Transform the Data
Do we know that this client spent the more money than client 66037? If not, how would we find out? Transform the data using the steps below to prepare it for analysis.

```python
# Create a column that calculates the subtotal for each line using the unit_price and the qty
ecommerce_df['sub_total'] = ecommerce_df['unit_price'] * ecommerce_df['qty']

sub_total_ecommerce_df = ecommerce_df[['client_id', 'order_id', 'unit_price', 'qty', 'sub_total']]
sub_total_ecommerce_df.head(10)
```

```python
# Create a column for shipping price.
# Assume a shipping price of $7 per pound for orders over 50 pounds and $10 per pound for items 50 pounds or under.

def calculate_shipping_price(unit_weight, qty):
    order_weight = unit_weight * qty
    if order_weight > 50:
        return 7 * order_weight
    else:
        return 10 * order_weight
    
ecommerce_df['shipping_price'] = ecommerce_df.apply(lambda row: calculate_shipping_price(row['unit_weight'], row['qty']), axis=1)

shipping_price_df = ecommerce_df[['client_id', 'order_id', 'unit_price', 'unit_weight', 'qty', 'sub_total', 'shipping_price']]
shipping_price_df.head(10)

```

```python
# Create a column for sales tax using the subtotal, shipping price and a 9.25% tax rate
ecommerce_df['sales_tax'] = (((ecommerce_df['sub_total']) + (ecommerce_df['shipping_price'])) * 0.0925).round(2)

sales_tax_df = ecommerce_df[['client_id', 'order_id', 'unit_price', 'unit_weight', 'qty', 'sub_total', 'shipping_price', 'sales_tax']]
sales_tax_df.head(10)
```

```python
# Create a column for the total price using the subtotal and the shipping price along with a sales tax of 9.25%
ecommerce_df['total_price'] = (ecommerce_df['sub_total'] + ecommerce_df['shipping_price'] + ecommerce_df['sales_tax'])

total_price_df = ecommerce_df[['client_id', 'order_id', 'unit_price', 'unit_weight', 'qty', 'sub_total', 'shipping_price', 'sales_tax', 'total_price']]
total_price_df.head(10)
```

```python
# Create a column for the cost of each line using unit cost, qty, and
# shipping price (assume the shipping cost is exactly what is charged to the client).
ecommerce_df['total_cost'] = ((ecommerce_df['unit_cost'] * ecommerce_df['qty'])) + ecommerce_df['shipping_price']

total_cost_df = ecommerce_df[['client_id', 'order_id', 'unit_price', 'unit_cost', 'unit_weight', 'qty', 'sub_total', 'shipping_price', 'sales_tax', 'total_price', 'total_cost']]
total_cost_df.head(10)
```

```python
# Create a column for the profit of each line using line cost and line price
ecommerce_df['profit'] = (ecommerce_df['total_price'] - ecommerce_df['total_cost']).round(2)

total_profit_df = ecommerce_df[['client_id', 'order_id', 'unit_price', 'unit_cost', 'unit_weight', 'qty', 'sub_total', 'shipping_price', 'sales_tax', 'total_price', 'total_cost', 'profit']]
total_profit_df.head(10)
```

## Part 4: Summarize and Analyze
Use the new columns with confirmed values to find the following information.

```python
# How much did each of the top 5 clients by quantity spend? Check your work from Part 1 for client ids.
for client in top_five_clients_list:
    total_spend = ecommerce_df[ecommerce_df['client_id'] == client]['total_price'].sum()
    print(f"Client {client} spent: ${total_spend:,.2f}")

    
        
        
```

```python
# Create a summary DataFrame showing the totals for the top 5 clients with the following information:
# total units purchased, total shipping price, total revenue, and total profit. 

# Initialize an empty list to store the summary data for each top client
summary_data = []

# Iterate over each client in the top five clients list
for client in top_five_clients_list:
    # Filter the DataFrame for the current client
    client_data = ecommerce_df[ecommerce_df['client_id'] == client]
    
    # Sum the relevant columns
    qty = client_data['qty'].sum()
    shipping_price = client_data['shipping_price'].sum()
    total_price = client_data['total_price'].sum()
    total_cost = client_data['total_cost'].sum()
    total_profit = client_data['profit'].sum()
    
    # Append the results as a dictionary
    summary_data.append({
        'client_id': client,
        'Units': qty,
        'Total Shipping Price': shipping_price,
        'Total Revenue': total_price,
        'Total Cost': total_cost,
        'Total Profit': total_profit
    })

# Create a new summary DataFrame from the list of dictionaries
top_five_summary_df = pd.DataFrame(summary_data)

# Display the summary DataFrame
top_five_summary_df

                                                                                                                                                            
```

```python
# Format the data and rename the columns to names suitable for presentation.

# Define the money columns. 
money_columns = ['Total Shipping Price', 'Total Revenue', 'Total Cost', 'Total Profit']

# Define a function that converts a dollar amount to millions.
def currency_format_millions(dollar_amount):
    return dollar_amount / 1000000

# Apply the currency_format_millions function to only the money columns. 
top_five_summary_df[money_columns] = top_five_summary_df[money_columns].apply(currency_format_millions)

# Rename the columns to reflect the change in the money format. 
top_five_summary_df.rename(columns={
    'client_id': 'Client ID',
    'Total Shipping Price': 'Shipping (Millions)',
    'Total Revenue': 'Total Revenue (Millions)',
    'Total Cost': 'Total Cost (Millions)',
    'Total Profit': 'Total Profit (Millions)',
    
}, inplace=True)

# Display the updated data frame.
top_five_in_millions_df = top_five_summary_df[['Client ID', 'Shipping (Millions)', 'Total Revenue (Millions)', 'Total Cost (Millions)', 'Total Profit (Millions)']]
top_five_in_millions_df

```

```python
# Sort the updated data by "Total Profit (millions)" from highest to lowest and assign the sort to a new DatFrame.
sorted_df = top_five_in_millions_df.sort_values(by='Total Profit (Millions)', ascending=False)

sorted_df
```
